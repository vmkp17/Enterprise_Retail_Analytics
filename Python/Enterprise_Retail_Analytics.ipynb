## 1. Business Background

The retail industry generates large volumes of customer transaction data every day.
````python
# ============================================================
# PHASE 3
# STEP 1 - Validate Customer Age
# ============================================================

print("=" * 70)
print("AGE VALIDATION")
print("=" * 70)

print("Minimum Age :", df["Age"].min())
print("Maximum Age :", df["Age"].max())

invalid_age = df[(df["Age"] < 18) | (df["Age"] > 100)]

print("\nInvalid Age Records :", len(invalid_age))

display(invalid_age)

# ============================================================
# STEP 2 - Purchase Amount Validation
# ============================================================

print("=" * 70)
print("PURCHASE AMOUNT VALIDATION")
print("=" * 70)

print("Minimum Purchase :", df["Purchase Amount (USD)"].min())
print("Maximum Purchase :", df["Purchase Amount (USD)"].max())

invalid_purchase = df[df["Purchase Amount (USD)"] <= 0]

print("\nInvalid Purchase Records :", len(invalid_purchase))

display(invalid_purchase)


# ============================================================
# STEP 3 - Review Rating Validation
# ============================================================

print("=" * 70)
print("REVIEW RATING VALIDATION")
print("=" * 70)

print("Minimum Rating :", df["Review Rating"].min())
print("Maximum Rating :", df["Review Rating"].max())

invalid_rating = df[
    (df["Review Rating"] < 1) |
    (df["Review Rating"] > 5)
]

print("\nInvalid Rating Records :", len(invalid_rating))

display(invalid_rating)


# ============================================================
# STEP 4 - Category Validation
# ============================================================

print("=" * 70)
print("CATEGORY VALIDATION")
print("=" * 70)

print(df["Category"].unique())

print("\nTotal Categories :", df["Category"].nunique())


# ============================================================
# STEP 5 - Season Validation
# ============================================================

print("=" * 70)
print("SEASON VALIDATION")
print("=" * 70)

print(df["Season"].unique())

print("\nTotal Seasons :", df["Season"].nunique())


# ============================================================
# STEP 6 - Payment Method Validation
# ============================================================
print("=" * 70)
print("PAYMENT METHOD VALIDATION")
print("=" * 70)

display(df["Payment Method"].value_counts())

# ============================================================
# STEP 7 - Shipping Type Validation
# ============================================================

print("=" * 70)
print("SHIPPING TYPE VALIDATION")
print("=" * 70)

display(df["Shipping Type"].value_counts())


# ============================================================
# FINAL VALIDATION REPORT
# ============================================================

validation_report = pd.DataFrame({
    "Validation Check": [
        "Missing Values",
        "Duplicate Rows",
        "Invalid Age",
        "Invalid Purchase Amount",
        "Invalid Review Rating"
    ],
    "Result": [
        df.isnull().sum().sum(),
        df.duplicated().sum(),
        len(invalid_age),
        len(invalid_purchase),
        len(invalid_rating)
    ],
    "Status": [
        "PASS" if df.isnull().sum().sum() == 0 else "FAIL",
        "PASS" if df.duplicated().sum() == 0 else "FAIL",
        "PASS" if len(invalid_age) == 0 else "FAIL",
        "PASS" if len(invalid_purchase) == 0 else "FAIL",
        "PASS" if len(invalid_rating) == 0 else "FAIL"
    ]
})

display(validation_report)



````

````python
# ============================================================
# PHASE 4
# STEP 1 - Create Age Group
# ============================================================

age_bins = [17, 25, 35, 45, 55, 70]

age_labels = [
    "18-25 Young Adult",
    "26-35 Adult",
    "36-45 Mid Adult",
    "46-55 Mature Adult",
    "56-70 Senior Adult"
]

df["Age Group"] = pd.cut(
    df["Age"],
    bins=age_bins,
    labels=age_labels
)

display(df[["Age", "Age Group"]].head())


# ============================================================
# STEP 2 - Spending Level
# ============================================================

df["Spending Level"] = pd.cut(
    df["Purchase Amount (USD)"],
    bins=[19, 40, 70, 100],
    labels=[
        "Low Spender",
        "Medium Spender",
        "High Spender"
    ]
)

display(df[
    [
        "Purchase Amount (USD)",
        "Spending Level"
    ]
].head())


# ============================================================
# STEP 3 - Loyalty Level
# ============================================================

df["Loyalty Level"] = pd.cut(
    df["Previous Purchases"],
    bins=[0, 15, 30, 50],
    labels=[
        "New Customer",
        "Regular Customer",
        "Loyal Customer"
    ]
)

display(df[
    [
        "Previous Purchases",
        "Loyalty Level"
    ]
].head())



# ============================================================
# STEP 4 - Rating Category
# ============================================================

df["Rating Category"] = pd.cut(
    df["Review Rating"],
    bins=[2.4, 3.5, 4.2, 5],
    labels=[
        "Average",
        "Good",
        "Excellent"
    ]
)

display(df[
    [
        "Review Rating",
        "Rating Category"
    ]
].head())


# ============================================================
# STEP 5 - Discount Status
# ============================================================

df["Discount Status"] = df["Discount Applied"]

display(df[
    [
        "Discount Applied",
        "Discount Status"
    ]
].head())

# ============================================================
# VERIFY NEW COLUMNS
# ============================================================

print("=" * 60)
print("NEW FEATURES CREATED")
print("=" * 60)

print(df.columns)

````

````python

# Pareto Analysis
pareto = (
    df.groupby("Item Purchased")["Purchase Amount (USD)"]
      .sum()
      .sort_values(ascending=False)
)

pareto_df = pareto.reset_index()
pareto_df.columns = ["Product", "Revenue"]

pareto_df["Cumulative Revenue"] = pareto_df["Revenue"].cumsum()
pareto_df["Cumulative %"] = (
    pareto_df["Cumulative Revenue"]
    / pareto_df["Revenue"].sum()
) * 100

display(pareto_df)

# ============================================================
# Customer Persona
# ============================================================

persona = (
    df.groupby(
        [
            "Age Group",
            "Gender",
            "Spending Level"
        ]
    )
    .size()
    .reset_index(name="Customers")
)

display(persona.sort_values("Customers", ascending=False))


# ============================================================
# Revenue by State and Category
# ============================================================

state_category = pd.pivot_table(
    df,
    values="Purchase Amount (USD)",
    index="Location",
    columns="Category",
    aggfunc="sum"
)

display(state_category)


# ============================================================
# Loyalty Matrix
# ============================================================

loyalty_matrix = pd.pivot_table(
    df,
    values="Purchase Amount (USD)",
    index="Loyalty Level",
    columns="Category",
    aggfunc="mean"
)

display(loyalty_matrix)

# ============================================================
# Executive Business Summary
# ============================================================

summary = pd.DataFrame({

    "Metric": [

        "Total Customers",

        "Total Revenue",

        "Average Purchase",

        "Average Rating",

        "Total States",

        "Total Categories",

        "Top Category"

    ],

    "Value": [

        df["Customer ID"].nunique(),

        df["Purchase Amount (USD)"].sum(),

        round(df["Purchase Amount (USD)"].mean(),2),

        round(df["Review Rating"].mean(),2),

        df["Location"].nunique(),

        df["Category"].nunique(),

        df["Category"].value_counts().idxmax()

    ]

})

display(summary)

````

````python
query = """
SELECT
    CASE
        WHEN "Purchase Amount (USD)" <= 40 THEN 'Low Spender'
        WHEN "Purchase Amount (USD)" <= 70 THEN 'Medium Spender'
        ELSE 'High Spender'
    END AS Spending_Level,

    COUNT(*) AS Total_Customers

FROM Retail_Data

GROUP BY Spending_Level

ORDER BY Total_Customers DESC;
"""
result = pd.read_sql(query, connection)
display(result)


query = """
SELECT

Category,

SUM("Purchase Amount (USD)") AS Revenue

FROM Retail_Data

GROUP BY Category

HAVING Revenue > 50000

ORDER BY Revenue DESC;
"""
result = pd.read_sql(query, connection)
display(result)


query = """
SELECT

Location,

SUM("Purchase Amount (USD)") AS Revenue

FROM Retail_Data

GROUP BY Location

ORDER BY Revenue DESC

LIMIT 5;
"""
result = pd.read_sql(query, connection)
display(result)

query = """
SELECT

"Customer ID",

"Purchase Amount (USD)",

ROW_NUMBER() OVER(
ORDER BY "Purchase Amount (USD)" DESC
) AS Customer_Rank

FROM Retail_Data;
"""
result = pd.read_sql(query, connection)
display(result.head(20))

query = """
SELECT

"Customer ID",

"Purchase Amount (USD)",

RANK() OVER(
ORDER BY "Purchase Amount (USD)" DESC
) AS Purchase_Rank

FROM Retail_Data;
"""
result = pd.read_sql(query, connection)
display(result.head(20))

````
